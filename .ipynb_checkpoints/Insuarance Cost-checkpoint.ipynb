## Linear Regression

```python
# from sklearn.linear_model import LinearRegression
```

```python
inputs=['age','bmi','children','smoker_val','sex_val','northeast','northwest','southeast','southwest']
x_train = x_train[inputs]
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
inputs=['age','bmi','children','smoker_val','sex_val','northeast','northwest','southeast','southwest']
x_test = x_test[inputs]
prediction = model.predict(x_test)
```

```python
# from sklearn.metrics import mean_squared_error
# import math
def loss(y_train, y_test, train_prediction, test_prediction, algo):
    rmse_train = np.sqrt(np.mean(np.square(y_train-train_prediction)))
    rmse_test = np.sqrt(np.mean(np.square(y_test-test_prediction)))
    print(algo)
    print('Train RMSE = ',rmse_train)
    print('Test RMSE = ',rmse_test)
```

```python
train_preds = model.predict(x_train)
loss(y_train, y_test, train_preds, prediction, "Linear Regression")
```

```python
weights = model.coef_
for i in range(len(inputs)):
    print(inputs[i], weights[i])
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
tree_model = dtr.fit(x_train, y_train)
```

```python
train_tree_prediction = tree_model.predict(x_train)
test_tree_prediction = tree_model.predict(x_test)
```

```python
test_tree_prediction.shape
```

```python
loss(y_train, y_test, train_tree_prediction, test_tree_prediction, "Decision Tree")
```

### As we can see, the model highly overfits the test data.

## Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
rf_model = rfr.fit(x_train, y_train)
train_tree_prediction_rf = rf_model.predict(x_train)
test_tree_prediction_rf = rf_model.predict(x_test)
loss(y_train, y_test, train_tree_prediction_rf, test_tree_prediction_rf, "Random Forest")
```

```python
def result(age, bmi, children, smoker, gender, region):
    if smoker=="yes":
        smoker_val=1
    else:
        smoker_val=0
    
    if region=="northeast":
        ne = 1
        nw = 0
        se = 0
        sw = 0
        
    elif region=='northwest':
        ne = 0
        nw = 1
        se = 0
        sw = 0
        
    elif region=='southeast':
        ne = 0
        nw = 0
        se = 1
        sw = 0
        
    elif region=='southwest':
        ne = 0
        nw = 0
        se = 0
        sw = 1
    
    if gender=='male':
        gender_val=1
    else:
        gender_val=0
    
    new = {
    'age': age,
    'bmi': bmi,
    'children': children,
    'smoker_val': smoker_val,
    'sex_val': gender_val,
    'northeast': ne,
    'northwest': nw,
    'southeast': se,
    'southwest': sw
    }
    new = pd.DataFrame([new])
    new[numeric_input] = standard_scaler.transform(new[numeric_input])
    return rf_model.predict(new)
```
